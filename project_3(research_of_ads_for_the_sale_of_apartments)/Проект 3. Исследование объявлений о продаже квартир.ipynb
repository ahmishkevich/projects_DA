#### Функция для проведения категоризации этажей (floor_group)

```python
# функция для проведения категоризации этажей
def floor_group(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
df['floor_category'] = df.apply(floor_group, axis=1)
```

#### Функция для подсчёта границ с учетом 1,5 межквартильного размаха (quartile_range)

```python
# функция для подсчёта границ с учетом 1,5 межквартильного размаха (сокращённо IQR)
def quartile_range(DataFrame, column):
    q1 = DataFrame[column].quantile(0.25)
    q3 = DataFrame[column].quantile(0.75)
    iqr = q3 - q1
    DataFrame_q = DataFrame.loc[(DataFrame[column] < q3 + 1.5*iqr) & (DataFrame[column] > q1 - 1.5*iqr), column]
    return DataFrame_q
```

### Заполнение пропусков

Сначала определим долю пропущенных значений среди всех данных. Для этого достаточно двух методов pandas:

```python
df.isna().mean() # подсчет доли пропущенных значений
```

- `ceiling_height` — около 38% данных о высоте потолков в метрах отсутствуют. Это может быть связанно с тем, что пользователи не всегда указывают данную информацию. Пропуски в таких колонках обычно заполняются характерными значениями: средним арифметическим или медианой. **В нашем случае лучше использовать медиану, ведь в данных могут быть выдяющиеся значения, которые сместят среднее**. Заполнять пропуски нулем или максимальным значением не стоит — такие значения нельзя назвать характерными;
- `floors_total` — меньше 1% данных о этажах в доме отсутствуют. На основе данных нельзя сделать точное предположение связанное с их отсутствием — это одноэтажные помещения или пропуск при заполнении. Было бы правильно обратиться к источнику информации, однако в нашем случае, в силу незначительности данных мы их отбросим;
- `living_area` и `locality_name` — имеют пропущенные значения меньше 10%. Возможно эти данные не успели пройти модерацию, т.к. данные о жилой площади квартиры и название населенного пункта являются обязательными при публикации объявлений. Такие данные не подлежат анализу, их необходимо исключить;
- `is_apartment` — около 88% данных отсутствуют. Скорее всего это связанно с тем, что при заполнении объявлений пользователи не указывали принадлежность к категории апартаментов. По умолчанию это означает, что недвижимость не является апартаментами;
- `kitchen_area` — примерно 9% данных о площади кухни отсутствуют, что также может указыувать на человеческий фактор возникновения пропусков. Заменим отсутствующие значения медианой;
- `balcony` — 48% данных о наличии балконов отсутствуют. Как и в случае с апартаментами, можно сделать предположение, что недвижимость не имеет балкона;
- Данные в столбцах `airports_nearest`, `city_centers_nearest`, `parks_around3000`, `ponds_around3000`, `parks_nearest` и `ponds_nearest` заполняются автоматически на основе картографических данных, исключая человеческий фактор. Здесь мы наблюдаем интересную аномалию: примерно в 77% случаев у недвижимости есть парк/пруд в пределах 3 км, но в 61-65% случаев нет информации об отдаленности. Было бы уместно обратиться к первоисточнику информации и ознакомиться с рассчетами формул, возможно в них есть ошибка. В нашем случае мы заменим эти данные медианой;
- `days_exposition` — в 13% случаев данные отсутствуют. Поскольку ошибка не может быть вызвана человеческим фактором, можно предположить, что она отражает объявления не прошедшие модерацию. Исключим их из выборки.


Исключим неопределенные значения из столбцов `locality_name`, `living_area`, `floors_total` и `days_exposition`.

```python
df.dropna(subset=['locality_name', 'living_area', 'floors_total', 'days_exposition'], inplace=True)
```

```python
df['kitchen_area'] = df['kitchen_area'].fillna(df.groupby('living_area')['kitchen_area'].transform('median'))
```

```python
df['is_apartment'] = df['is_apartment'].fillna(False) # замена пропусков
```

Пропуски столбца `balcony` заменим на значение 0.

```python
df['balcony'] = df['balcony'].fillna(0)
```
