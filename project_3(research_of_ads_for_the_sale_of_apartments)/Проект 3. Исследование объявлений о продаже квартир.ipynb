### Анализ редких и выбивающихся значений

Убрать редкие и выбивающиеся значения. Описать, какие особенности были обнаружены.
Редкие и выбивающиеся значения являются выбросами — точками данных, которые лежат вдали от обычного распределения данных и приводит к тому, что воздействуют и искажают общее распределение данных. В процессе анализа они были обнаруженны в столбцах `last_price`, `total_area`, `kitchen_area`, `living_area`, `days_exposition`, `ceiling_height` и `rooms`. Избавимся от них, применив функцию подсчета границ `quartile_range` на основе «ящика с усами».

```python
# функция для подсчёта границ с учетом 1,5 межквартильного размаха (сокращённо IQR)
def quartile_range(DataFrame, column):
    q1 = DataFrame[column].quantile(0.25)
    q3 = DataFrame[column].quantile(0.75)
    iqr = q3 - q1
    DataFrame_q = DataFrame.loc[(DataFrame[column] < q3 + 1.5*iqr) & (DataFrame[column] > q1 - 1.5*iqr), column]
    return DataFrame_q
```

Применим функцию к столбцам с выбросами.

```python
for row in ['last_price', 'total_area', 'kitchen_area', 'living_area', 'days_exposition', 'ceiling_height', 'rooms']:
    df[row] = quartile_range(df, row)
```
