#### Функция первичного анализа данных (hist_maker_plot)

```python
df['tmp'] = 1 # служит вместо счетчика строк (для подсчета процента)
```

```python
# функция построения графика
def hist_maker_plot(df, groupby, top, sort): # столбец группировки, категория группировки, размер топа, порядок графика
       
    # группировка и создание сводной таблицы
    all_groupby = df.groupby(groupby)['tmp'].sum().reset_index().sort_values(by='tmp', ascending=sort)[:top]
    
    # построение графика
    plot = sns.barplot(x=all_groupby[groupby], y=all_groupby['tmp'], color='blue', alpha=0.5, ec='black')
    
    # добавление процентов на график
    total = sum(df['tmp']) # расчет общих продаж региона
    
    for p in plot.patches: # цикл для каждого столбца графика
        width = p.get_width() # ширина
        height = p.get_height() # высота
        x, y = p.get_xy()
        percentage = '{:.1f}%'.format(100 * p.get_height() / total) # расчет процентов
        
        plot.annotate(percentage, (x + width / 2, y + height * 1.01), ha='center') # размещение процентов


    plt.ylabel('Объекты общественного питания')
    plot.set_xlabel(groupby)
    plot.set_title('Распределение объектов общественного питания по {}'.format(groupby))
    
    print('За 100% принято считать {} объектов(та) общественного питания'.format(total))
```

#### Функция отделения информации об улице из адреса (street_fix)

```python
street_tags = []

# функция отделения информации об улице из адреса
def street_fix(address):
    for address_part in address.split(', '):
        for street_tag in street_tags:
            if address_part.lower().find(street_tag)!= -1:
                return address_part
```

### Анализ зависимости числа посадочных мест от числа объектов общественного питания с характерным сетевым распространением

Рассмотрим, что характерно для сетевых заведений: много заведений с небольшим числом посадочных мест в каждом или мало заведений с большим количеством посадочных мест?

```python
chain_yes = df.query('chain == "да"').groupby('object_name').agg({'number':'mean', 'tmp': 'count'}).reset_index() # сводная таблица сетевых заведений
chain_yes.columns = ['object_name', 'mean_number', 'tmp'] # наименование заведений, количество посадочных мест в среднем по сетевым заведениям, число сетевых заведений

# построение графика
plt.figure(figsize=(10,5))
plt.scatter(chain_yes['mean_number'], chain_yes['tmp'], color='#8678da', alpha=0.5, ec='black')
plt.xlabel('mean_number')
plt.ylabel('object_name')

# отметки выбросов
plt.axhline(np.percentile(chain_yes['tmp'], 90), linestyle='--', color='black')
plt.axvline(np.percentile(chain_yes['mean_number'], 90), linestyle='--', color='black')

plt.show()

print('В 90% сетей не больше {} посадочных мест и {} заведений в среднем'.format(int(np.percentile(chain_yes['mean_number'], 90)) , int(np.percentile(chain_yes['tmp'], 90))))
```

На основе графика можно сделать вывод, что чем меньше сеть, тем больше количество посадочных мест, чем больше сеть, тем меньше посадочных мест.
